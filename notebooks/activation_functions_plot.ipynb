```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
%matplotlib inline

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 15
fig_size[1] = 3
# plt.rcParams["figure.figsize"] = fig_size

x = np.arange(-6,6, 0.1)
rx = np.arange(-1,1, 0.1)
f, (p1,p2,p3)= plt.subplots(1,3)

#Sigmoid
y = 1/(1+np.exp(-x))
p1.set_title('Sigmoid')
p1.grid(True)
p1.plot(x,y, color='b')

#TanH
y = 2/(1+np.exp(-2*x))-1
p2.set_title('TanH')
p2.grid(True)
p2.plot(x,y,  color='red')

#ReLU
y = np.maximum(0,rx)
p3.set_title('ReLU')
p3.grid(True)
p3.plot(rx,y, color='black')
```
